```python
w = tf.constant(1.)
x = tf.constant(2.)
y = x*w
```

```python
with tf.GradientTape() as tape:
    tape.watch([w])
    y2 = x*w
    
```

```python
# y 位在Gradient裡面 無法求解
grad1 = tape.gradient(y,[w])
grad1
```

```python
# y 位在Gradient裡面 無法求解
grad2 = tape.gradient(y2,[w]) 
grad2
```

```python
#一般梯度只能求解一次 若要兩次 要加上persistent = True
with tf.GradientTape(persistent = True) as tape:
    tape.watch([w])
```

```python
#MSE Gradient
x = tf.random.normal([2,4])
#4維->3維 
w = tf.random.normal([4,3])
b =tf.zeros([3])
y = tf.constant([2,0])
```

```python
with tf.GradientTape() as tape:
    tape.watch([w,b])
    prob = tf.nn.softmax(x@w + b, axis =1)
    loss = tf.reduce_mean(tf.losses.MSE(tf.one_hot(y, dept =3), prob))
```
